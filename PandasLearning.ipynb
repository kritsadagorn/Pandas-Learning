# Sorting / Describeing Data

```python
df.sort_values(['Type 1', 'HP'], ascending=[1,0])
```

# Making changes to the data

```python
# df["NEW"] = df['Attack'] + df['Defense'] # add new column named NEW
# df = df.drop(columns=['NEW']) # <----- drop the column

df['NEW'] = df.iloc[:, 4:10].sum(axis=1)  # add NEW column of summary of column 4-9 in Horizontal axis

cols = list(df.columns.values)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df.head(5)
```

# Saving Data

# Filtering Data

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

# new_df.to_csv('new_df.csv') # save filtered file

new_df.reset_index(drop=True, inplace=True)
new_df
```

```python
import re

df.loc[df['Name'].str.contains('pi[a-z]*', regex=True, flags=re.I)]
```

# Conditional Changes

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1' ] = 'Flamer'
df
```

# Working with large amounts of data

```python
new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('pokemon_data.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()

    new_df = pd.concat([new_df, results])
```
